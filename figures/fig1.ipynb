```python
import numpy as np
import matplotlib.pyplot as plt
import quante as qt
from quante.basicfun import load_hdf5, plt_style_use
plt_style_use("quante", svg=True)
```

```python
def fig_a(L, γ, ax:plt.Axes):
    filename = "sigmaM_TFI.h5"
    group = f"L={L}/γ={γ:.3f}"
    eig_NH = load_hdf5(filename=filename, group=group, data="eigs_NH")
    qt.measure.plot_level_spacing_distribution(eig_NH, ax, st_dists=["poisson-c","ginibre"])
    # load data from gas position distribution
    beta = 1.5
    s_range, func = load_hdf5(filename="gas.h5", group=f"beta_{beta}", data=["s_range", "func"])
    ax.plot(s_range, func, 'k--', linewidth=2, label=f'Coulomb gas, β={beta}')
    

def fig_b(L, γ, ax:plt.Axes):
    filename = "sigmaM_TFI.h5"
    group = f"L={L}/γ={γ:.3f}"
    eig_Liou = load_hdf5(filename=filename, group=group, data="eigs_Liou")
    qt.measure.plot_level_spacing_distribution(eig_Liou, ax, st_dists=["poisson-c", "ginibre"])
```

```python
fig = plt.figure(figsize=(11, 5))

# Energy spacing distribution
## nonH
ax = fig.add_subplot(2, 3, 1)
fig_a(L=9, γ=1., ax=ax)

## open
ax = fig.add_subplot(2, 3, 4)
fig_b(L=9, γ=1., ax=ax)

# Complex spacing ration
## nonH
ax = fig.add_subplot(2, 3, 2)
## open
ax = fig.add_subplot(2, 3, 5)

# Singular value spacing distribution
## nonH
ax = fig.add_subplot(2, 3, 3)
### inset
## open
ax = fig.add_subplot(2, 3, 6)
### inset
```
